```python
import pandas as pd
import json
```

```python
def read_datastore(file_name):
    try:
        with open(file_name, 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        print(f'Error reading datastore: File not found: {file_name}')
        return None
    except json.JSONDecodeError:
        print(f'Error reading datastore: Invalid JSON format in file: {file_name}')
        return None
```

```python
data = read_datastore('data.json')['data']
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
total_distance = df.distance.sum()
total_distance
```

```python
dist_by_sport_type = df.groupby(by='sport_type').distance.sum()
dist_by_sport_type
```

```python
for cat, dist in dist_by_sport_type.items():
    print(f'Percentage of total distance in category {cat}: {dist*100 / total_distance}%')
```
