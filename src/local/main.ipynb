```python
import pandas as pd
from numpy import float64 as np_float64, int64 as np_int64
import json
```

```python
def read_datastore(file_name):
    try:
        with open(file_name, 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        print(f'Error reading datastore: File not found: {file_name}')
        return None
    except json.JSONDecodeError:
        print(f'Error reading datastore: Invalid JSON format in file: {file_name}')
        return None
```

```python
def get_weekly(column, target_df=df):
    if "start_date_dt" not in df.columns:
        df["start_date_dt"] = pd.to_datetime(df["start_date"])

    return (target_df
                .groupby(pd.Grouper(key="start_date_dt", freq="W-MON", label="left", closed="left"))[column]
                .sum()
                .rename(index=lambda ts: int(ts.timestamp()) * 1000)
                .to_dict())
```

```python
d = get_weekly("distance")
d[list(d.keys())[0]]
```

```python
d = {}

for sport in df["sport_type"].unique():
    d[sport] = get_weekly("distance", df[df["sport_type"] == sport])

sum_ = 0

for sport, data in d.items():
    first_key = list(data.keys())[0]
    print(first_key)
    sum_ += data[first_key]

sum_
```

```python
def get_weekly_by_sport(column):
    d = {}

    for sport in df["sport_type"].unique():
        d[sport] = get_weekly(column, df[df["sport_type"] == sport])

    return d
```
